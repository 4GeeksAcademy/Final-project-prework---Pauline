```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon
import pandas as pd
df = pd.read_csv('assets/real_estate.csv', sep=';')
cinturon_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
subconjunto_cinturonsur = df[df['level5'].isin(cinturon_sur)]

dicc_coord = subconjunto_cinturonsur.set_index('Unnamed: 0')[['latitude', 'longitude', 'level5']].to_dict(orient='index')

col_city = {
    'Alcorcón': 'lightgreen',
    'Getafe': 'lightblue',
    'Fuenlabrada': 'purple',
    'Leganés': 'gray'}

for yes, values in dicc_coord.items():
    latitude = float(values['latitude'].replace(',', '.'))
    longitude = float(values['longitude'].replace(',', '.'))
    level5 = values['level5']
    icono = AwesomeIcon(name='star',  marker_color=col_city.get(level5, 'blue'))
    city_mark = Marker(location=(latitude, longitude), icon=icono)
    map.add_layer(city_mark)

map

```
